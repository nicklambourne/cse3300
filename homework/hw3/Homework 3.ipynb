### Part (a)

Nodal Delay = $d_{nodal} = d_{queue} + d_{trans} + d_{prop} + d_{proc}$

Ignore queuing delay, propagation delay, processing delay (leaving only transmission delay). There are two nodes.

Total delay = $2 \cdot d_{trans}$

$d_{trans} = \frac{L}{R}$, $L =$ packet length, $R =$ bandwidth.

Total delay = $2 \cdot \frac{64,000 bits}{100,000bps}$

Total delay = $1.28\, seconds$

### Part (b)

#### Section (1)

Segment size = $1,000bytes$. Each segment has header size = $40 bits$. 

Number of segments = $64,000/8,000 = 8\, segments$

Therefore total segment size $= 1,000bytes + 40 bits = 8,040bits$

$d_{trans} = \frac{L}{R}$, $L =8,040$ (packet length), $R = 100,000bps$ (bandwidth).

Total delay = $2 \cdot \frac{8,040 bits}{100,000bps} \cdot 1\, segment + \frac{8,040 bits}{100,000bps} \cdot 7\, segments$

Total delay = $0.7236\, seconds$

#### Section (2)

Segment size = $100bytes$. Each segment has header size = $40 bits$.

Number of segments = $64,000bits/800bits = 80\, segments$.

Therefore total segment size $= 100bytes + 40 bits = 840bits$

$d_{trans} = \frac{L}{R}$, $L =840$ (packet length), $R = 100,000bps$ (bandwidth).

Total delay = $2 \cdot\frac{840 bits}{100,000bps} \cdot 1\, segment + \frac{840 bits}{100,000bps} \cdot 79\, segments$.

Total delay = $0.6804\, seconds$
